### Resumo Estatístico das variáveis

#### **Função que calcula o Resumo Estatístico para variáveis númericas**

```python
import pandas as pd

def res(series):
    # Calcula estatísticas descritivas usando agg
    resumo = series.agg(['count', 'mean','median', 'std', 'min', 'max' ]).round(2)

    # Renomeiar a coluna 'median' para 'median (50%)'
    resumo = resumo.rename(index={'median': 'median (50%)'})


    # Calcula percentis usando quantile
    percentis = series.quantile([0.25, 0.5, 0.75]).round(2)

    # Combina os resultados em um DataFrame único
    resumo = pd.concat([resumo, percentis.rename({0.25: '25%', 0.5: '50%', 0.75: '75%'})])

    return resumo

```

#### **Função que retorna valor único**

```python
import pandas as pd

def uniq(series):
    # Calcula a contagem dos valores únicos usando value_counts
    contagem = series.value_counts().reset_index()

    # Renomeia as colunas para tornar o resultado mais claro
    contagem.columns = ['Valor_Único', 'Frequência']

    # Adiciona uma coluna de proporção em porcentagem
    contagem['Proporção (%)'] = (contagem['Frequência'] / contagem['Frequência'].sum()) * 100

    # Arredonda a coluna de proporção para 4 casas decimais
    contagem['Proporção (%)'] = contagem['Proporção (%)'].round(4)

    # Formata a coluna de proporção como porcentagem
    contagem['Proporção (%)'] = contagem['Proporção (%)'].map('{:.4f}%'.format)

    return contagem

```

##### **Criar tabelas com o tipo de comida com o maior número de pedidos para cada Religião Predominante**

<b>1.Agrupar por religião: encontrar as linhas com o maior número de pedidos e assim encontrar r_cuisine com maior nº de pedidos</b>
<br>
Será feito o agrupamento por <code>Religiao_predominante</code> e  <code>r_cuisine1</code>. Em seguida, será somado as colunas <code>n_pedidos</code> e <code>n_pedidos_porcentage</code> para cada <code>r_cuisine1</code>.

```python
# Agrupar por "Religiao_predominante" e "r_cuisine1" e somar as colunas 'n_pedidos' e 'n_pedidos_porcentagem' para cada "r_cuisine1"
resultados_agrupados = resumo_por_religiao.groupby(['Religiao_predominante', 'r_cuisine1']).agg({
    'n_pedidos': 'sum',
    'n_pedidos_porcentagem': 'sum'
}).reset_index()
resultados_agrupados
```

<b>2.Tabela com o tipo de comida com o maior número de pedidos para cada Religião Predominante</b>
<br>
A partir da tabela anterior, será aplicado um filtro para selecionar as linhas com o maior <code>n_pedidos</code> para cada <code>Religiao_predominante</code>.

```python
# Encontrar os índices dos maiores valores de "n_pedidos_porcentagem" para cada "Religiao_predominante"
indices_maior_porcentagem = resultados_agrupados.groupby('Religiao_predominante')['n_pedidos_porcentagem'].idxmax()

# Selecionar os resultados com maior "n_pedidos_porcentagem" para cada "Religiao_predominante"
resultados_maior_porcentagem = resultados_agrupados.loc[indices_maior_porcentagem]

# Exibir o resultado
resultados_maior_porcentagem
```

<b>3. Tabela Final com os 4 tipos de comida com o mais pedidos para cada Religião Predominante</b>
<br>
A partir da tabela <code>resultados_agrupados</code>, será aplicado um filtro para selecionar os 4 maiores valores de <code>n_pedidos</code> para cada <code>Religiao_predominante</code>.
<br>
A tabela final será usada para criar o gráfico. 

```python
# Selecionar os 4 maiores resultados de "n_pedidos_porcentagem" para cada "Religiao_predominante"
resultados_top4 = (resultados_agrupados.groupby('Religiao_predominante')
                   .apply(lambda group: group.nlargest(4, 'n_pedidos_porcentagem'))
                   .reset_index(drop=True))
resultados_top4                   
```

# Regra de associação 

## Introdução à Regras de Associação
As regras de associação ajudam a descobrir relacionamentos entre itens de uma base de dados. Essas regras possuem alguns elementos importantes como:<br>
**Antecedentes**: Conjunto de itens que precisam estar presentes na transação para que a regra se aplique.<br>
**Consequentes**: Conjunto de itens que é previsto pela regra. Em outras palavras, se uma transação contém o(s) antecedente(s), a regra prevê que ela também conterá o(s) consequente(s) com uma certa probabilidade.
Supondo que o antecedente seja leite o consequente seja pão.<br>



<img src="../Imagens/Regra_de_Associação/leite.implica.em.pao.png" width="50%" >


**Antecedent support:** Proporção de transações no seu conjunto de dados que contém o(s) antecedente(s). <br>
<img src="../Imagens/Regra_de_Associação/antecedent_support.png" width="70%"> <br>
Por exemplo, supondo que antecedent support é igual a 0.14. Significa que 14% das transações da base de dados possuem leite.

**Consequent support:** Proporção de transações no seu conjunto de dados que contém o(s) consequente(s).<br>
<img src='../Imagens/Regra_de_Associação/consequent_support1.png' width="70%"> <br>
Por exemplo, supondo que consequent support é igual a 0.21. Significa que 21% das transações da base de dados possuem pão.

**Support:** Proporção de transações no seu conjunto de dados que contêm ambos os itens, tanto o antecedente quanto o consequente. Ou seja, é o número de transações para as quais uma regra faz a predição correta (utilidade). Suporte está associado a significância de estatística da regra.
<br>
Esta métrica serve para eliminar regras pouco interessantes, ou, eliminar conjunto de itens que não atendem ao critério mínimo. 
<br>
<img src="../Imagens/Regra_de_Associação/support.png" width="70%">
<br>
Por exemplo, sendo o antecedente pão e o consequente leite. Se o support for igual a 0.10, significa que 10% das transações contém pão e leite juntos.

**Confidence:** número de transações que a regra prediz corretamente entre as transações para as quais a regra se aplica (certeza). Ou seja, esta medida define a probabilidade de ocorrência do consequente no carrinho dado que o carrinho já possui os antecedentes. Esta métrica está associada a acurácia ou precisão.
<br>
<img src="../Imagens/Regra_de_Associação/confidence.png" width="70%">
<br>
Por exemplo, se o confidence for igual a 0.76, significa que há 76% de chance de uma transação que tem pão também ter leite.

**Lift:** mede o grau em que dois itens são dependentes um do outro, em comparação com o que seria esperado se fossem independentes. 
<br>
Um valor de elevação de 1 indica que os dois itens são independentes, um valor maior que 1 indica uma associação positiva, enquanto um valor menor que 1 indica uma associação negativa.
<br>
Por exemplo, se lift for igual a 2.1 , significa que a chance de encontrar pão e leite juntos é 2.1 vezes maior do que encontrar cada um individualmente.


## Regra de associação 1: usando categoria de restaurante 

```python
import pandas as pd 
import os

#Definir Diretorio
os.chdir("C:/0.Projetos/3.Analise_Exploratoria_Zomato")

#Carregar tabela
dados9=pd.read_csv("Outros_datasets/dados9_tabela_para_analise_descritiva.csv")
tab1=dados9
```

### Algoritmo Apriori 1

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
#1º PASSO
# Criar uma tabela com as colunas "f_id" e "r_cuisine1"
df1 = tab1[['user_id', 'r_cuisine1']]

#2º PASSO
# Contagem de ocorrências por combinação de user_id e r_cuisine
counts1 = df1.groupby(['user_id', 'r_cuisine1']).size().reset_index(name='count')
# Adicionar prefixo "cuisine" às colunas antes de espalhar
counts1['r_cuisine1'] = 'cuisine_' + counts1['r_cuisine1'].astype(str)

#3º PASSO
# Espalhar as contagens para diferentes colunas
pivot_table1 = counts1.pivot(index='user_id', columns='r_cuisine1', values='count').fillna(0)

#4º PASSO
#Tranformar a tabela em operadores booleanos
reg_ass1=pivot_table1
reg_ass1 = reg_ass1.astype(bool)

#5º PASSO
# Aplicar algoritmo Apriori para encontrar itens frequentes
frequent_items1 = apriori(reg_ass1, min_support=0.005, use_colnames=True)

#6º PASSO
# Gerar regras de associação
rules1 = association_rules(frequent_items1, metric="confidence", min_threshold=0.5)
```

```python
# Exibir as regras de associação
print(rules1)
```

### Algoritmo FP-growth 1

```python
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules
# Supondo que você já tenha seu DataFrame e tenha calculado os itemsets frequentes com FP-growth
frequent_items_fp1 = fpgrowth(reg_ass1, min_support=0.01, use_colnames=True)

# Gerar regras de associação
rules_fp1 = association_rules(frequent_items_fp1, metric="confidence", min_threshold=0.6)

# Exibir as regras
print(rules_fp1)
```
